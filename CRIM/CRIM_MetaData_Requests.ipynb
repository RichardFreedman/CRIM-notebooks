```python
def get_ema_for_observation_id(obs_id):
    # get Obs_1_ema
    my_ema_mei_dictionary = dict()
    url = "https://crimproject.org/data/observations/{}/".format(obs_id)
    response = requests.get(url)
    Obs_json = response.json()
    
    # Obs_ema = Obs_json["ema"]
    
    my_ema_mei_dictionary["id"]=Obs_json["id"]
    my_ema_mei_dictionary["musical type"]=Obs_json["musical_type"]
    my_ema_mei_dictionary["int"]=Obs_json["mt_fg_int"]
    my_ema_mei_dictionary["tint"]=Obs_json["mt_fg_tint"]
    my_ema_mei_dictionary["ema"]=Obs_json["ema"]
    my_ema_mei_dictionary["mei"]=Obs_json["piece"]["mei_links"][0]
    my_ema_mei_dictionary["pdf"]=Obs_json["piece"]["pdf_links"][0]
   
    
    # Obs_piece = Obs_json["piece"]
    # Obs_mei = Obs_piece["mei_links"]
    
    print(f'Got: {obs_id}')
    
    # return {"ema":Obs_ema,"mei":Obs_mei}
    
    return my_ema_mei_dictionary
```

```python
#This is Rich's attempt to gather voice names for a given OBS
#Gets a particular OBS, then lists the PieceID, the OBS_ID and then all of the fields that contain voices.

def get_voices_for_observation_id(obs_id):
    
    my_voices_dict = dict()
    url = "https://crimproject.org/data/observations/{}/".format(obs_id)
    response = requests.get(url)
    Obs_json = response.json()
    
    my_voices_dict["piece_id"]=Obs_json["piece"]["piece_id"]
    my_voices_dict["id"]=Obs_json["id"]
    my_voices_dict["mt_cf_voices"]=Obs_json["mt_cf_voices"]
    my_voices_dict["mt_sog_voices"]=Obs_json["mt_sog_voices"]
    my_voices_dict["mt_csog_voices"]=Obs_json["mt_csog_voices"]
    my_voices_dict["mt_cd_voices"]=Obs_json["mt_cd_voices"]
    my_voices_dict["mt_fg_voices"]=Obs_json["mt_fg_voices"]
    my_voices_dict["mt_pe_voices"]=Obs_json["mt_pe_voices"]
    my_voices_dict["mt_id_voices"]=Obs_json["mt_id_voices"]
    my_voices_dict["mt_nid_voices"]=Obs_json["mt_nid_voices"]
    my_voices_dict["mt_hr_voices"]=Obs_json["mt_hr_voices"]
    my_voices_dict["mt_int_voices"]=Obs_json["mt_int_voices"]
    my_voices_dict["mt_cad_cantizans"]=Obs_json["mt_cad_cantizans"]
    my_voices_dict["mt_cad_tenorizans"]=Obs_json["mt_cad_tenorizans"]
    my_voices_dict["mt_cad_dtv"]=Obs_json["mt_cad_dtv"]
    
                                                   
#for observ in range(1,11):
    #call_list_output = get_voices_for_observation_id(observ)
    
    # the print command simply puts the output in the notebook terminal.  
    #Later we will put it in the List of Dicts.
    
   # print(call_list_output)
   # obs_data_list.append(call_list_output)
    
    # Obs_piece = Obs_json["piece"]
    # Obs_mei = Obs_piece["mei_links"]
    
    print(f'Got: {obs_id}')
    
    # return {"ema":Obs_ema,"mei":Obs_mei}
    
    return my_voices_dict
```
